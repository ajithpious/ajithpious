# Dimensionality Reduction

```python
from sklearn.decomposition import PCA
pca=PCA(0.95)
X_transformed=pca.fit_transform(x_feature_selected)
variance=pca.explained_variance_ratio_
print(variance*100)
```

```python
fig,ax=plt.subplots(figsize=(15,4))
variance=list(map(lambda x:round(x,2)*100,variance))
print(variance)
ax.bar(x=range(1,25),height=variance)
ax.set_xticks(range(1,24))
ax.set_xlabel("Principal Components")
ax.set_ylabel("Variance")
ax.set_title("Pricipal components Variance")
for index,value in enumerate(variance):
  ax.text(index,value,"      "+str(value))
plt.show()
```

## Random Forest Classifier

```python
# Finding the optimal hyper parameters
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
params = {'criterion':['gini'],
          'n_estimators':range(50,100,5),
          'max_depth':range(3,10),
          'min_samples_leaf':[3,4],
          'min_samples_split':[6,7], 
          'random_state':[123],
          'n_jobs':[-1]}

model_rf = RandomizedSearchCV(rf, param_distributions=params, n_jobs=-1,cv=10,n_iter=150)
model_rf.fit(xtrain,ytrain)
print("Best Hyper Parameters:\n",model_rf.best_params_)
```

```python
# Finding the optimum value of Max-Depth
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
rf1=RandomForestClassifier()
params1 = {'criterion':['gini'],
          'n_estimators':[100,110,115],
          'max_depth':range(3,10),
          # 'min_samples_leaf':[1,2,3],
          'min_samples_split':[6,7], 
          'random_state':[12],
          'n_jobs':[-1]}
model_rf1 = RandomizedSearchCV(rf1, param_distributions=params1, n_jobs=-1,cv=10,n_iter=150)
model_rf1.fit(xtrain,ytrain)
print("Best Hyper Parameters:\n",model_rf1.best_params_)
```

## SVM

```python
# import SVC classifier
from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV

svc=SVC() 

params = {'kernel':['linear', 'poly'],
          # 'gamma':[0.001, 0.01, 0.1, 1, 10, 100],
          'C':[0.001, 0.01, 0.1, 1, 10, 100,1000],
          'degree':[2, 3, 4],
          'random_state':[123]
          }

model_svc = RandomizedSearchCV(svc, param_distributions=params, n_jobs=-1,cv=10,n_iter=15)
model_svc.fit(xtrain,ytrain)
```

## ROC

```python
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve,auc
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC
```

```python
y=label_binarize(ytrain,classes=[1,2,3,4,5])
```

```python
yroc=label_binarize(y_test,classes=[1,2,3,4,5])
```

```python
yroc.shape
```

```python
classifier=OneVsRestClassifier(SVC(kernel= 'poly',  degree= 3, C= 10,probability=True)).fit(xtrain, y)
```

```python
y_pred=classifier.decision_function(x_eval_transformed)
y_pred
```

```python
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(5):
    fpr[i], tpr[i], thres = roc_curve(yroc[:, i],y_pred[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
```

```python
thres
```

```python
fig,ax=plt.subplots(figsize=(20,10))

for i in range(5):
    ax.plot(fpr[i], tpr[i], label='ROC curve of class {0} (area = {1:0.2f})'
                                 ''.format(i+1, roc_auc[i]))
ax.plot([0, 1], [0, 1], color="navy", lw=2, linestyle="--",label="Random Classifier")
ax.set_xlabel("False Positive Rate")
ax.set_ylabel("True Positive Rate")
ax.set_title("ROC for SVM")
ax.legend()
```
